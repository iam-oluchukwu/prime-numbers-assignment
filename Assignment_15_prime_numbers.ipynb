# Assignment 15 - Prime Numbers
## Objective
This program prints all prime numbers between 1 and a given number (n).

## Method
- Take input from user
- Use nested for loop
- Store prime numbers in a list

```python
n = int(input("Enter a number: "))

prime_numbers = []

for num in range(2, n + 1):
    is_prime = True
    
    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break
    
    if is_prime:
        prime_numbers.append(num)

print(prime_numbers)
```
